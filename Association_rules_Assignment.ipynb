# Apriori algorithm

```python
#performing apriori algorithm
frequent_itemsets = apriori(books,min_support=0.01,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets1 = apriori(books,min_support=0.1,use_colnames=True)
frequent_itemsets1
```

```python
frequent_itemsets2= apriori(books,min_support=0.4,use_colnames=True)
frequent_itemsets2
```

```python
rules = association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
rule_1=rules[ (rules['support'] >= 0.1 )&(rules['confidence'] >= 0.7) ]
```

```python
rule_2=rules[(rules['support'] >= 0.1)&(rules['confidence'] >= 0.8)]
```

```python
rule_3=rules[(rules['support'] >=0.2)&(rules['confidence'] >= 0.6)]
```

```python
sns.scatterplot(x='support',y='confidence',data=rule_1)
#plt.xlabel(support)
#plt.ylabel(confidence)
#plt.show()
```

```python
sns.scatterplot(x='support',y='confidence',data=rule_2,alpha=1)
```

```python
sns.scatterplot(x='support',y='confidence',data=rule_3)
```

# Association rules for movies data set

```python
movies=pd.read_csv('my_movies.csv')
movies.head(1)
```

```python
movies.drop(['V1','V2','V3','V4','V5'],axis=1,inplace=True)
```

# since the data is not in transaction format we are not performing transaction encoding

```python
#te=TransactionEncoder()
#te_ary=te.fit(movies).transform(movies)
#te_ary.shape
```

```python
#te_ary
```

```python
#df=pd.DataFrame(te_ary,columns=)
```

```python
frequent = apriori(movies,min_support=0.2,use_colnames=True)
frequent
```

```python
frequent1 = apriori(movies,min_support=0.3,use_colnames=True)
frequent1
```

```python
frequent2 = apriori(movies,min_support=0.1,use_colnames=True)
frequent2
```

```python
movie_rules = association_rules(frequent,metric='lift',min_threshold=0.7)
movie_rules
```

```python
movie_rule1=movie_rules[ (movie_rules['support'] >= 0.1 )&(movie_rules['confidence'] >= 0.7) ]
movie_rule1
```

```python
movie_rule2=movie_rules[ (movie_rules['support'] >= 0.3 )&(movie_rules['confidence'] >= 0.8) ]
movie_rule2
```

```python
movie_rule3=movie_rules[ (movie_rules['support'] >= 0.3 )&(movie_rules['confidence'] >= 0.9) ]
movie_rule3
```

```python
sns.scatterplot(x='support',y='confidence',data=movie_rule1)
```

```python
sns.scatterplot(x='support',y='confidence',data=movie_rule2)
```
